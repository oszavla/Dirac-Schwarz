# Komponen Tak-Nol Simbol Christiffel, Tensor Riemann, dan Tensor Ricci Metrik Schwarzschild

```python
import sympy as sp

def hitung_tensor_schwarzschild():
    # 1. Definisi Simbol dan Fungsi
    # Menggunakan c untuk kecepatan cahaya sesuai SI
    t, r, theta, phi = sp.symbols('t r theta phi')
    c = sp.symbols('c', positive=True)

    # alpha(r) dan beta(r) bergantung secara spasial pada r
    alpha = sp.Function('alpha')(r)
    beta = sp.Function('beta')(r)

    coords = [t, r, theta, phi]
    coord_names = ['t', 'r', 'theta', 'phi']

    # 2. Definisi Metrik g_{mu,nu}
    g = sp.zeros(4, 4)
    g[0, 0] = -c**2 * sp.exp(2*alpha)
    g[1, 1] = sp.exp(2*beta)
    g[2, 2] = r**2
    g[3, 3] = r**2 * sp.sin(theta)**2

    # Metrik Invers g^{mu,nu}
    g_inv = g.inv()

    # 3. Menghitung Simbol Christoffel (\Gamma^\lambda_{\mu\nu})
    print("="*50)
    print(r'SIMBOL CHRISTOFFEL TAK NOL (\Gamma^\lambda_{\mu\nu})')
    print("="*50)
    Gamma = sp.MutableDenseNDimArray.zeros(4, 4, 4)
    for lam in range(4):
        for mu in range(4):
            for nu in range(4):
                term = 0
                for rho in range(4):
                    term += 0.5 * g_inv[lam, rho] * (
                        sp.diff(g[rho, nu], coords[mu]) +
                        sp.diff(g[rho, mu], coords[nu]) -
                        sp.diff(g[mu, nu], coords[rho])
                    )
                # Menyimpan dan menyederhanakan
                Gamma[lam, mu, nu] = sp.simplify(term)

                # Hanya menampilkan yang tidak nol
                if Gamma[lam, mu, nu] != 0:
                    print(f"Gamma^{coord_names[lam]}_{coord_names[mu]}{coord_names[nu]} = {Gamma[lam, mu, nu]}")

    # 4. Menghitung Tensor Riemann (R^\rho_{\sigma\mu\nu})
    print("\n" + "="*50)
    print(r'TENSOR RIEMANN TAK NOL (R^\rho_{\sigma\mu\nu})')
    print("="*50)
    Riemann = sp.MutableDenseNDimArray.zeros(4, 4, 4, 4)
    for rho in range(4):
        for sig in range(4):
            for mu in range(4):
                for nu in range(4):
                    # Turunan parsial
                    t1 = sp.diff(Gamma[rho, nu, sig], coords[mu])
                    t2 = sp.diff(Gamma[rho, mu, sig], coords[nu])

                    # Kontraksi indeks dummy lambda
                    t3 = sum(Gamma[rho, mu, lam] * Gamma[lam, nu, sig] for lam in range(4))
                    t4 = sum(Gamma[rho, nu, lam] * Gamma[lam, mu, sig] for lam in range(4))

                    R_comp = sp.simplify(t1 - t2 + t3 - t4)
                    Riemann[rho, sig, mu, nu] = R_comp

                    # Menampilkan komponen tak nol
                    if R_comp != 0:
                        print(f"R^{coord_names[rho]}_{coord_names[sig]}{coord_names[mu]}{coord_names[nu]} = {R_comp}")

    # 5. Menghitung Tensor Ricci (R_{\mu\nu})
    print("\n" + "="*50)
    print(r'TENSOR RICCI (R_{\mu\nu})')
    print("="*50)
    Ricci = sp.zeros(4, 4)
    for mu in range(4):
        for nu in range(4):
            # Kontraksi Tensor Riemann atas indeks rho: R_{\mu\nu} = R^\rho_{\mu\rho\nu}
            val = sum(Riemann[rho, mu, rho, nu] for rho in range(4))
            Ricci[mu, nu] = sp.simplify(val)

            # Menampilkan SEMUA komponen (baik nol maupun tak nol) sesuai permintaan
            if Ricci[mu, nu] != 0:
                print(f"R_{coord_names[mu]}{coord_names[nu]} [TAK NOL] = {Ricci[mu, nu]}")
            else:
                print(f"R_{coord_names[mu]}{coord_names[nu]} = 0")

if __name__ == "__main__":
    hitung_tensor_schwarzschild()
```

```python
import sympy as sp

def hitung_dan_cetak_latex():
    # 1. Definisi Simbol dan Fungsi
    t, r, theta, phi = sp.symbols('t r theta phi')
    c = sp.symbols('c', positive=True)

    alpha = sp.Function('alpha')(r)
    beta = sp.Function('beta')(r)

    coords = [t, r, theta, phi]
    # Penamaan untuk LaTeX
    coord_names = ['t', 'r', '\\theta', '\\phi']

    # 2. Definisi Metrik g_{mu,nu}
    g = sp.zeros(4, 4)
    g[0, 0] = -c**2 * sp.exp(2*alpha)
    g[1, 1] = sp.exp(2*beta)
    g[2, 2] = r**2
    g[3, 3] = r**2 * sp.sin(theta)**2

    g_inv = g.inv()

    print("% --- Salin kode di bawah ini ke dalam dokumen LaTeX Anda ---")
    print("\\begin{align*}")

    # 3. Menghitung Simbol Christoffel (\Gamma^\lambda_{\mu\nu})
    print("% Simbol Christoffel")
    Gamma = sp.MutableDenseNDimArray.zeros(4, 4, 4)
    for lam in range(4):
        for mu in range(4):
            for nu in range(4):
                term = 0
                for rho in range(4):
                    term += 0.5 * g_inv[lam, rho] * (
                        sp.diff(g[rho, nu], coords[mu]) +
                        sp.diff(g[rho, mu], coords[nu]) -
                        sp.diff(g[mu, nu], coords[rho])
                    )
                Gamma[lam, mu, nu] = sp.simplify(term)

                if Gamma[lam, mu, nu] != 0:
                    # Format ke sintaks LaTeX
                    latex_expr = sp.latex(Gamma[lam, mu, nu])
                    print(f"\\Gamma^{{{coord_names[lam]}}}_{{{coord_names[mu]}{coord_names[nu]}}} &= {latex_expr} \\\\")

    # 4. Menghitung Tensor Riemann (R^\rho_{\sigma\mu\nu})
    print("\\\\ % Tensor Riemann")
    Riemann = sp.MutableDenseNDimArray.zeros(4, 4, 4, 4)
    for rho in range(4):
        for sig in range(4):
            for mu in range(4):
                for nu in range(4):
                    t1 = sp.diff(Gamma[rho, nu, sig], coords[mu])
                    t2 = sp.diff(Gamma[rho, mu, sig], coords[nu])
                    t3 = sum(Gamma[rho, mu, lam] * Gamma[lam, nu, sig] for lam in range(4))
                    t4 = sum(Gamma[rho, nu, lam] * Gamma[lam, mu, sig] for lam in range(4))

                    R_comp = sp.simplify(t1 - t2 + t3 - t4)
                    Riemann[rho, sig, mu, nu] = R_comp

                    if R_comp != 0:
                        latex_expr = sp.latex(R_comp)
                        print(f"R^{{{coord_names[rho]}}}_{{{coord_names[sig]}{coord_names[mu]}{coord_names[nu]}}} &= {latex_expr} \\\\")

    # 5. Menghitung Tensor Ricci (R_{\mu\nu})
    print("\\\\ % Tensor Ricci")
    Ricci = sp.zeros(4, 4)
    for mu in range(4):
        for nu in range(4):
            val = sum(Riemann[rho, mu, rho, nu] for rho in range(4))
            Ricci[mu, nu] = sp.simplify(val)

            # Mencetak semua komponen (0 maupun tak nol)
            latex_expr = sp.latex(Ricci[mu, nu])
            print(f"R_{{{coord_names[mu]}{coord_names[nu]}}} &= {latex_expr} \\\\")

    print("\\end{align*}")

if __name__ == "__main__":
    hitung_dan_cetak_latex()
```

# Keseluruhan Komponen Simbol Christofel, Tensor Riemann, Tensor Ricci, Skalar Ricci, dan Tensor Einstein pada Metrik Schwarzschild

```python
import sympy as sp

def hitung_tensor_gr():
    # 1. Definisi Simbol dan Koordinat
    # Menggunakan representasi LaTeX untuk nama variabel agar output langsung rapi
    t, r, theta, phi = sp.symbols('t r \\theta \\phi')
    coords = [t, r, theta, phi]
    indeks_str = ['t', 'r', '\\theta', '\\phi']

    # Definisi fungsi untuk komponen metrik
    alpha = sp.Function('\\alpha')(r)
    beta = sp.Function('\\beta')(r)

    # 2. Definisi Tensor Metrik Kovarian (g_mu_nu)
    g = sp.zeros(4, 4)
    g[0, 0] = -sp.exp(2 * alpha)
    g[1, 1] = sp.exp(2 * beta)
    g[2, 2] = r**2
    g[3, 3] = r**2 * sp.sin(theta)**2

    # Invers Tensor Metrik Kontravarian (g^mu_nu)
    g_inv = g.inv()

    print("% =========================================")
    print("% KODE LATEX HASIL PERHITUNGAN TENSOR GR")
    print("% =========================================\n")

    # 3. Menghitung Simbol Christoffel
    # Gamma^rho_{mu nu} = 1/2 * g^{rho lam} * (d_mu g_{nu lam} + d_nu g_{mu lam} - d_lam g_{mu nu})
    Gamma = sp.MutableDenseNDimArray.zeros(4, 4, 4)
    for rho in range(4):
        for mu in range(4):
            for nu in range(4):
                term = 0
                for lam in range(4):
                    term += 0.5 * g_inv[rho, lam] * (
                        sp.diff(g[nu, lam], coords[mu]) +
                        sp.diff(g[mu, lam], coords[nu]) -
                        sp.diff(g[mu, nu], coords[lam])
                    )
                Gamma[rho, mu, nu] = sp.simplify(term)

    print("\\subsection*{Simbol Christoffel ($\\Gamma^\\rho_{\\mu\\nu}$)}")
    print("\\begin{align*}")
    for rho in range(4):
        for mu in range(4):
            for nu in range(4):
                val = sp.simplify(Gamma[rho, mu, nu])
                print(f"\\Gamma^{{{indeks_str[rho]}}}_{{{indeks_str[mu]}{indeks_str[nu]}}} &= {sp.latex(val)} \\\\")
    print("\\end{align*}\n")


    # 4. Menghitung Tensor Riemann
    # Konvensi: R^rho_{sigma mu nu} = d_mu Gamma^rho_{nu sigma} - d_nu Gamma^rho_{mu sigma} + Gamma^rho_{mu lam}Gamma^lam_{nu sigma} - Gamma^rho_{nu lam}Gamma^lam_{mu sigma}
    Riemann = sp.MutableDenseNDimArray.zeros(4, 4, 4, 4)
    for rho in range(4):
        for sigma in range(4):
            for mu in range(4):
                for nu in range(4):
                    term = sp.diff(Gamma[rho, nu, sigma], coords[mu]) - sp.diff(Gamma[rho, mu, sigma], coords[nu])
                    for lam in range(4):
                        term += Gamma[rho, mu, lam] * Gamma[lam, nu, sigma] - Gamma[rho, nu, lam] * Gamma[lam, mu, sigma]
                    Riemann[rho, sigma, mu, nu] = sp.simplify(term)

    print("\\subsection*{Tensor Riemann ($R^\\rho_{\\sigma\\mu\\nu}$)}")
    print("\\begin{align*}")
    for rho in range(4):
        for sigma in range(4):
            for mu in range(4):
                for nu in range(4):
                    val = sp.simplify(Riemann[rho, sigma, mu, nu])
                    # Karena Tensor Riemann memiliki 256 komponen, kita memecahnya agar compiler LaTeX tidak error overfull hbox
                    print(f"R^{{{indeks_str[rho]}}}_{{{indeks_str[sigma]}{indeks_str[mu]}{indeks_str[nu]}}} &= {sp.latex(val)} \\\\")
    print("\\end{align*}\n")


    # 5. Menghitung Tensor Ricci
    # R_{mu nu} = R^rho_{mu rho nu}
    Ricci = sp.zeros(4, 4)
    for mu in range(4):
        for nu in range(4):
            term = 0
            for rho in range(4):
                term += Riemann[rho, mu, rho, nu]
            Ricci[mu, nu] = sp.simplify(term)

    print("\\subsection*{Tensor Ricci ($R_{\\mu\\nu}$)}")
    print("\\begin{align*}")
    for mu in range(4):
        for nu in range(4):
            val = sp.simplify(Ricci[mu, nu])
            print(f"R_{{{indeks_str[mu]}{indeks_str[nu]}}} &= {sp.latex(val)} \\\\")
    print("\\end{align*}\n")


    # 6. Menghitung Skalar Ricci
    # R = g^{mu nu} R_{mu nu}
    R_scalar = 0
    for mu in range(4):
        for nu in range(4):
            R_scalar += g_inv[mu, nu] * Ricci[mu, nu]
    R_scalar = sp.simplify(R_scalar)

    print("\\subsection*{Skalar Ricci ($R$)}")
    print("\\begin{equation*}")
    print(f"R = {sp.latex(R_scalar)}")
    print("\\end{equation*}\n")


    # 7. Menghitung Tensor Einstein
    # G_{mu nu} = R_{mu nu} - 1/2 * g_{mu nu} * R
    Einstein = sp.zeros(4, 4)
    for mu in range(4):
        for nu in range(4):
            Einstein[mu, nu] = sp.simplify(Ricci[mu, nu] - 0.5 * g[mu, nu] * R_scalar)

    print("\\subsection*{Tensor Einstein ($G_{\\mu\\nu}$)}")
    print("\\begin{align*}")
    for mu in range(4):
        for nu in range(4):
            val = sp.nsimplify(Einstein[mu, nu])
            print(f"G_{{{indeks_str[mu]}{indeks_str[nu]}}} &= {sp.latex(val)} \\\\")
    print("\\end{align*}\n")

if __name__ == "__main__":
    hitung_tensor_gr()
```
